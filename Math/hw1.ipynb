```python
import matplotlib.pyplot as plt # as (keyword) it's nickname 
import numpy as np

def f(x):
    return 2*x + 1 # kx + b

x = np.linspace(-3,3,10)
y = f(x) # f() - function которая принимает аргумент

plt.plot(x, y)
plt.title('График линейной функции')
plt.xlabel('x')
plt.ylabel('y')
plt.grid() # сетка
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
# x = []
# help(np.linspace) (function) # Help on function linspace in module numpy:
# list # (method) Help on built-in function append:
```
